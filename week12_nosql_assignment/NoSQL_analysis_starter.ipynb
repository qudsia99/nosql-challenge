### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest


# sorting by hygiene, using $match operator due to multiple conditions to be applied
query_1 = {"$match": {'scores.Hygiene': {"$eq":0}}}


# When querying with $group, we must follow up with _id, as it is present by default in all mongoDB's,
# since we are grouping, we create a new field 'count' that counts each document using $sum operator,
# with a value of +1, per document
group_2 = {'$group': {'_id': '$LocalAuthorityName','count': {'$sum': 1}}}

# Proceeding to sort by the count of each grouped LocalAuthorityName
sort = {'$sort': {'count': -1}}

# Creating a pipeline
pipeline = [query_1,group_2,sort]

hygiene = list(est.aggregate(pipeline))

# Print the number of documents in the result
count = list(hygiene)
print(len(count))

# Print the first 10 results
pprint(hygiene[0:10])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_df = pd.DataFrame(hygiene)

# Display the number of rows in the DataFrame
hygiene_df.count()

# Display the first 10 rows of the DataFrame
hygiene_df.head(10)
```
